# **Decision Trees**

This is an supervised learning algorithm which is used to solve both regression and classification problems. It is a predictive modelling approach that gives a graphical representation of decisions and every possible outcome because of that decision. As the name suggests , this approach is used to take important decisions based on previous data. 

The below diagram explains a typical structure of decisions trees.


image.png

The **ROOT NODE** is the main node. Consider this to the bark of a tree . Now, the **Branches** consist of a rule . Based on these different outcomes are produced. These outcomes are called **LEAF NODES**.

The major challenge in decision trees lies in attribute selection for root node at each level. This is done by following two metrics:

1. **Entropy (Information Gain)** : Entropy is measure of randomness in data points.Information Gain is chnage in Entropy. So, lower entropy means greater chnage and hence more information Gain. It is best suited for partitions with small counts but many distinct values.

2. **Gini Index** :  Measure of how often a randomly chosen element would be incorrectly identified.Attribute with lower Gini index is best. Also, it is best suited for larger partitions. 

### I've taken IRIS DATASET to explain decision trees.

### Step 1: Importing Libraries

```
import pandas as pd
import sklearn.datasets as datasets # this has many datasets,check its documentation for more info.
iris = datasets.load_iris() #To load iris dataset
```

### Step 2: Getting Data

```
data = pd.DataFrame(iris.data,columns = iris.feature_names)
#TIP : Always print whatever you do , to keep checking the results. 
#I've commented many commands, remove # before them and run them to check what they print. This will help you in understanding better. 
#data.head()
y = iris.target
#print(y)
```

###Step 3: Making Decision Tree

```
from sklearn import tree # for regression import DecisionTreeRegressor
dtr = tree.DecisionTreeClassifier()
dtr.fit(data,y)
#Decision Tree is created. The values displayed below are default values of this. You can change it according to your preferences.
```

###Step 4: Visualizing Decision Tree

```
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(20, 20)) #Set image size before hand else it gives very small image
tree.plot_tree(dtr,ax=ax,feature_names=iris.feature_names,filled=True,class_names=iris.target_names)
```

### Step 5: Understanding the decision tree

To understand this let's use some examples.

Case 1:

Take sepal_length = 2.5 ,sepal_width = 1,petal_length = 1.5 ,petal_width = 2 . Root node question is petal length<=2.45 which is True and hence, class is setosa.

Case 2: 

Take sepal_length = 2.5 ,sepal_width = 1, petal_length = 2.46,petal_width = 2 . Root node question is petal length<=2.45 which is False and hence, we move to next question petal width <= 1.75 , which is also false. So next question is petal length<=4.85 which is True.  Now the question come, sepal length<=5.95 which is also True and hence the class is versicolor.

To test the predictions , I'm using predict formula below.



### Step 6: Testing

```
case_one_pred = int(dtr.predict([[2.5, 1, 1.5, 2]]))
case_two_pred = int(dtr.predict([[2.5, 1, 2.46, 2]]))
print(f"The first case most likely belongs a {iris.target_names[case_one_pred]} flower.")
print(f"The second case most likely belongs a {iris.target_names[case_two_pred]} flower.")
```
